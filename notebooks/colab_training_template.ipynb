# Wildlife Detection YOLOv8n Training
This notebook trains YOLOv8n on the custom 14-class wildlife dataset.

```python
!pip install -U ultralytics
```

```python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
model.train(data='data/dataset.yaml', epochs=100, imgsz=640, batch=16)
```

```python
model = YOLO('runs/detect/train/weights/best.pt')
metrics = model.val(data='data/dataset.yaml')
print(metrics.box.map50, metrics.box.map, metrics.box.mp, metrics.box.mr)
```

```python
model.export(format='tflite', imgsz=640)
```
